Scraping data of cars in Delhi from olx:

```python
# typing cars in search field
s_field=driver.find_element(By.XPATH,"//form[@class='rui-1Azjs _20jkh']/fieldset/div/input")
s_field.send_keys("used cars in delhi")
time.sleep(2)
#Clicking search button
s_btn=driver.find_element(By.XPATH,"//div[@class='_3b3oR']")
s_btn.click()
time.sleep(3)
```

```python
# Clicking load button to see more cars
for i in range(0,25):
    if i>25:
        break
    else:
        load_btn=driver.find_element(By.XPATH,"//button[@class='rui-39-wj rui-3evoE rui-1JPTg']")
        load_btn.click()
        time.sleep(5)
    i=i+1
```

```python
# scraping data
brandname1=[]
model1=[]
variant1=[]
manufacturingyear1=[]
fuel1=[]
owner1=[]
location1=[]
price1=[]
```

```python
 #Scraping car manufacturing year
try:
    for i in driver.find_elements(By.XPATH,"//span[@class='_2TVI3']"):
        manufacturingyear1.append(i.text)
except NoSuchElementException:
    manufacturingyear1.append("----")
    time.sleep(5)
urls=[]
for i in range(0,1):
    if i>1:
        break
    a=driver.find_elements(By.XPATH,"//ul[@class='rl3f9 _3mXOU']/li/a")
    for i in a:
        urls.append(i.get_attribute('href'))
   
    for i in urls:
        driver.get(i)
        time.sleep(2)
        # scraping car brands
        try:
            for i in driver.find_elements(By.XPATH,"//div[@class='_35xN1']"):
                brandname1.append(i.text)
        except NoSuchElementException:
            brandname1.append("----")

        # scraping car model
        try:
            for i in driver.find_elements(By.XPATH,"//div[@class='_3tLee']"):
                model1.append(i.text)
        except NoSuchElementException:
            model1.append("----")
        # scraping car variant
        try:
            for i in driver.find_elements(By.XPATH,"//div[@class='aOxkz']/div[3]"):
                variant1.append(i.text)
        except NoSuchElementException:
            variant1.append("----")
       
        # scraping car fuel type
        try:
            for i in driver.find_elements(By.XPATH,"//div[@class='aOxkz']/div[1]"):
                fuel1.append(i.text)
        except NoSuchElementException:
            fuel1.append("----")
        # scraping car owner no.
        try:
            for i in driver.find_elements(By.XPATH,"//div[@class='_1l939']/div[1]/div[2]/div[2]"):
                owner1.append(i.text)
        except NoSuchElementException:
            owner1.append("----")
        # scraping car location
        try:
            for i in driver.find_elements(By.XPATH,"//div[@class='_1l939']/div[2]/div[2]/div[2]"):
                location1.append(i.text)
        except NoSuchElementException:
            location1.append("----")
        # scraping price of the car
        try:
            for i in driver.find_elements(By.XPATH,"//div[@class='_3FkyT']"):
                price1.append(i.text)
        except NoSuchElementException:
            price1.append("----")
```

```python
# making dataframe of scraped data from olx
df1=pd.DataFrame(list(zip(brandname1,model1,variant1,manufacturingyear1,fuel1,owner1,location1,price1)),columns=["BrandName","Model","Variant","ManufacturingYear","Fuel type","Owner","Location","Price"])
df1
```

```python
df1[['Brand','ModelName','mn_year']]=pd.DataFrame(df1.BrandName.str.split(' ',2).tolist(),columns=['Brand','ModelName','mn_year'])
```

```python
df1[['Year','Kilometer']]=pd.DataFrame(df1.ManufacturingYear.str.split(' - ',1).tolist(),columns=['Year','Kilometer'])
df1
```

```python
df1.drop(['BrandName','ManufacturingYear'],inplace=True,axis=1)
```

```python
df1=df1[['Brand','ModelName','Model','Variant','Fuel type','Kilometer','Owner','Location','Year','mn_year','Price']]
df1
```
